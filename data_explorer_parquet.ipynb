```python
import os
import pyarrow.parquet as pq

parquet_folder = 'parquets'
files = [f for f in os.listdir(parquet_folder) if f.endswith('.parquet')]

for fname in files:
    parquet_path = os.path.join(parquet_folder, fname)
    try:
        table = pq.read_table(parquet_path)
        columns = set(table.schema.names)
        has_timestamp = "Timestamp" in columns
        has_value = "Value" in columns
        print(f"\nFile: {fname}")
        print("Schema:", table.schema)
        if has_timestamp and has_value:
            df = table.to_pandas()
            print(df[['Timestamp', 'Value']].head())
        else:
            print(f"Missing columns in {fname}: ", end="")
            if not has_timestamp: print("Timestamp ", end="")
            if not has_value: print("Value ", end="")
            print()
    except Exception as e:
        print(f"Failed to read {fname}: {e}")
```
